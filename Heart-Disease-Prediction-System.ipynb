 **Preprocessing**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('heart_statlog_cleveland_hungary_final.csv')
df.head(10)
```

```python
df=df.drop_duplicates()
df.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

scaler.fit(df[['resting bp s']])
df['resting bp s'] = scaler.transform(df[['resting bp s']])

scaler.fit(df[['cholesterol']])
df['cholesterol'] = scaler.transform(df[['cholesterol']])

scaler.fit(df[['max heart rate']])
df['max heart rate'] = scaler.transform(df[['max heart rate']])

df.head(10)
```

**Visualisation**

```python
x1 = df.loc[df.sex==1, 'sex']
x2 = df.loc[df.sex==0, 'sex']
gender=['Female','Male']
xpos = np.arange(len(gender))
plt.hist(x1, color='g', label='Male')
plt.hist(x2, color='b', label='Female')
plt.gca().set(title='Heart Disease Distribution on men and women', ylabel='Count',xlabel='Gender')
plt.xticks(xpos,gender)
plt.legend();
```

```python
disease = len(np.array(df.loc[df.target==1, 'target']))
normal = len(np.array(df.loc[df.target==0, 'target']))
exp_vals = [disease,normal]
exp_labels = ["Heart Disease","Normal"]
plt.pie(exp_vals,labels=exp_labels,radius=1.5,autopct='%1.1f%%')
```

```python
c1 = 0;
c2 = 0;
c3 = 0;
c4 = 0;
df.rename(columns = {'chest pain type':'chest_pain_type'}, inplace = True)
for a,b in zip(df.target, df.chest_pain_type): 
    if(a==1):
      if(b==1):
        c1 = c1 +1;
      elif(b==2):
       c2 = c2 +1;
      elif(b==3):
       c3 = c3 +1;
      elif(b==4):
       c4 = c4 +1;
print("Type of chest pain in patients having heart diseases")
print("Typical angina", c1)
print("Atypical angina", c2)
print("Non-anginal pain", c3)
print("Asymptomatic", c4)
print("\n\n")

exp_vals = [c1,c2,c3,c4]
exp_labels = ["Typical angina","Atypical angina","Non-anginal pain","Asymptomatic"]
plt.pie(exp_vals,labels=exp_labels,radius=1.5,autopct='%1.1f%%')
```

```python
c1 = 0;
c2 = 0;
df.rename(columns = {'exercise angina':'exercise_angina'}, inplace = True)
for a,b in zip(df.target, df.exercise_angina): 
    if(a==1):
      if(b==1):
        c1 = c1 +1;
      elif(b==0):
       c2 = c2 +1;

print("Patients having exercise induced angina -", c1)
print("Patients not having exercise induced angina -", c2)
print("\n\n")

exp_vals = [c1,c2]
exp_labels = ["Exercise induced angina","No angina"]
plt.pie(exp_vals,labels=exp_labels,radius=1.5,autopct='%1.1f%%')
```

```python
c1 = 0;
c2 = 0;
c3 = 0;
df.rename(columns = {'resting ecg':'resting_ecg'}, inplace = True)
for a,b in zip(df.target, df.resting_ecg): 
    if(a==1):
      if(b==0):
        c1 = c1 +1;
      elif(b==1):
       c2 = c2 +1;
      elif(b==2):
       c3 = c3 +1;
print("Resting electrocardiogram in patients having heart diseases")
print("Normal", c1)
print("ST-T wave abnormality", c2)
print("left ventricular hypertrophy", c3)
print("\n\n")

exp_vals = [c1,c2,c3]
exp_labels = ["Normal","ST-T wave abnormality","left ventricular hypertrophy"]
plt.pie(exp_vals,labels=exp_labels,radius=1.5,autopct='%1.1f%%')
```

```python
c1 = 0;
c2 = 0;
c3 = 0;
df.rename(columns = {'ST slope':'ST_slope'}, inplace = True)
for a,b in zip(df.target, df.ST_slope): 
    if(a==1):
      if(b==0):
        c1 = c1 +1;
      elif(b==1):
       c2 = c2 +1;
      elif(b==2):
       c3 = c3 +1;
print("The slope of the peak exercise ST segment in patients having heart diseases")
print("Upsloping", c1)
print("Flat", c2)
print("Downsloping", c3)
print("\n\n")

exp_vals = [c1,c2,c3]
exp_labels = ["Upsloping","Flat","Downsloping"]
plt.pie(exp_vals,labels=exp_labels,radius=1.5,autopct='%1.1f%%',explode=[0.1,0.1,0.1])
```

**Training** **and** **Testing**

```python
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2)
```

```python
x_train.shape, x_test.shape
```

```python
# Logistic Regression
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(solver='lbfgs', max_iter=1000)
lr.fit(x_train, y_train)
x_train_prediction = lr.predict(x_train)
training_data_accuracy = accuracy_score(x_train_prediction, y_train)
print("Accuracy on training data: ", training_data_accuracy)
```

```python
x_test_prediction = lr.predict(x_test)
test_data_accuracy = accuracy_score(x_test_prediction, y_test)
print("Accuracy on test data: ", test_data_accuracy)
```
